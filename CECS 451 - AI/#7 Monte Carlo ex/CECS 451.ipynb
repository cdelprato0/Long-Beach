```python
import random
import math
import matplotlib.pyplot as plt


def calcPi(n):
    # holds all points
    points = []
    
    #randomly assigns x and y with numbers between 0-1
    for i in range(n):
        # creates the "sqaure of points"
        x = random.uniform(0, 1)
        y = random.uniform(0, 1)
        points.append((x,y))

    #holds the distances
    distances = []
    #points outside cirlce
    red = []
    #for each point in all the points
    for point in points:
        #find the distance from the origin to the point
        distance = math.sqrt((0 - point[0])**2 + (0 - point[1])**2)
        # if the distnace is less than or equal to 1, add the point to inner circle
        if(distance <= 1):
            distances.append(point)
        else:
            #add red points to outside
            red.append(point)
        
    # determine experimental pie by using monte carlo
    exPi = (4 * len(distances)) / n
    # calculates the percent error
    error = abs(exPi - math.pi)/math.pi *100
    
    # per instructions on hw, plot the red and bluee dots
    if(n == 10**4):
        plt.figure(figsize=(8,8))
        zip_poi = zip(*distances)
        plt.scatter(*zip_poi,color='b',s=1)
        zip_red = zip(*red)
        plt.scatter(*zip_red,color='r',s=1)

        plt
    
    return exPi, error

n3 = 10**3
n4 = 10**4
n5 = 10**5
n6 = 10**6

exPi, error = calcPi(n3)
format = "n = 10 ^ 3 pi = %s error = %.4f %%"
print(format % (exPi, error))
exPi, error = calcPi(n4)
format = "n = 10 ^ 4 pi = %s error = %.4f %%"
print(format % (exPi, error))
exPi, error = calcPi(n5)
format = "n = 10 ^ 5 pi = %s error = %.4f %%"
print(format % (exPi, error))
exPi, error = calcPi(n6)
format = "n = 10 ^ 6 pi = %s error = %.4f %%"
print(format % (exPi, error))

```
